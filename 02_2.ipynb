<h1>맷플롯립으로 그래프 그리기</h1>

맷플롯립은 과학 생태계 표준 그래프 패키지
대부분의 그래프를 그릴 수 있음

```python
import matplotlib.pyplot as plt
```

<h3>1. 선 그래프 그리기</h3>

```python
plt.plot([1, 2, 3, 4, 5,], [1, 4, 9, 16, 25])
plt.show()
```

<h3>2. 산점도 그리기</h3>

```python
plt.scatter([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])
plt.show()
```

<h3>3. 넘파이 배열로 산점도 그리기</h3>

```python
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.scatter(x, y)
plt.show()
```
